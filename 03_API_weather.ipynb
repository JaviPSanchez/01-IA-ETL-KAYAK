```python
df_coord = pd.read_csv('assets/top_35_coords_final.csv', index_col=0)
df_coord
```

```python
cities_coord_lat = []
cities_coord_lon = []
for i, rows in df_coord.iterrows():
    # print(rows)
    cities_coord_lat.append(rows['lat'])
    cities_coord_lon.append(rows['lon'])
display(cities_coord_lat[17])
display(cities_coord_lon[0:10])
```

```python
def get_meteo(lat, lon):
    k = 0
    for i, j in zip(lat, lon):
        response = requests.get(f"https://api.openweathermap.org/data/2.5/weather?lat={i}&lon={j}&appid={KEY_API_WEATHER}")
        data_coords = response.json()
        print(f"{k}: {data_coords}")
        k += 1
        yield data_coords

top_35_temperature = []
for data in get_meteo(cities_coord_lat, cities_coord_lon):
    # print(data['main']['temp'])
    if  data['cod'] == 400:
        top_35_temperature.append('No data')
    else:
        top_35_temperature.append(data['main']['temp'])
```

```python
top_35_temperature_final = pd.DataFrame(top_35_temperature)
top_35_temperature_final.columns =['temp_K']
top_35_temperature_final
```

```python
top_35_temperature_final.to_csv('assets/top_35_temperature.csv')
```
