```python
df_coord = pd.read_csv('assets/top_35_coords_final.csv', index_col=0)
df_coord.head()
```

```python
cities_coord_lat = []
cities_coord_lon = []
for i, rows in df_coord.iterrows():
    # print(rows)
    cities_coord_lat.append(rows['lat'])
    cities_coord_lon.append(rows['lon'])
display(cities_coord_lat[0:5])
display(cities_coord_lon[0:5])
```

```python
lat_1 = cities_coord_lat[0:2]
lon_1 = cities_coord_lon[0:2]
display(lat_1)
display(lon_1)
```

```python
def get_meteo(lat, lon):
    for data_lat in lat:
        for data_lon in lon:
            response = requests.get(f"https://api.openweathermap.org/data/2.5/weather?lat={data_lat}&lon={data_lon}&appid={KEY_API_WEATHER}")
            data_coords = response.json()
            yield data_coords
```

```python
get_meteo(lat_1, lon_1)
```

```python
top_35_temperature = []
for data in get_meteo(lat_1, lon_1):
    print(data['main']['temp'])
    top_35_temperature.append(data['main']['temp'])
```

```python
top_35_temperature_final = pd.DataFrame(top_35_temperature)
top_35_temperature_final
```
