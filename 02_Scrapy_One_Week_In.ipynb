```python
import os
import scrapy
import pandas as pd
from scrapy.crawler import CrawlerProcess
```

```python
class OneWeekSpider(scrapy.Spider):

    name = "oneweekin"
    start_urls = ['https://one-week-in.com/35-cities-to-visit-in-france/']


    def parse(self, response):
        results = response.css('div.entry-content')
        for p in results:
            yield {
                'city': p.css('h2 ::text').getall(),
                'text': p.css('::text').getall()
            }
        
```

```python
filename = "top_35_cities_to_visit_in_france.json"

# Declare a new CrawlerProcess with some settings
process = CrawlerProcess(settings = {
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)',
    "FEEDS": {
        filename : {"format": "json"},
    }
})

# Start the crawling using the spider you defined above
process.crawl(OneWeekSpider)
process.start()
```
