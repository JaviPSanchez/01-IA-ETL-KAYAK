```python
import pandas as pd
import requests
```

```python
df = pd.read_csv('assets/top_35_clean.csv', index_col=0)
df.head(40)
```

```python
response = requests.get("https://nominatim.openstreetmap.org/")
response
```

```python
cities_list = []
for i, rows in df.iterrows():
    cities_list.append(rows[0])
cities_list
```

```python
def get_coordinates(input_list):
    for city in input_list:
        gps = requests.get(f"https://nominatim.openstreetmap.org/search?city={city}&format=json&limit=1")
        data = gps.json()
        yield data
```

```python
get_coordinates(cities_list)
```

```python
top_35_coords_lat = []
top_35_coords_lon = []
for data in get_coordinates(cities_list):
    if data ==[]:
        # print('No data')
        top_35_coords_lat.append('No data')
        top_35_coords_lon.append('No data')
    else:
        # print(data[0]['lat'], data[0]['lon'])
        top_35_coords_lat.append(data[0]['lat'])
        top_35_coords_lon.append(data[0]['lon'])
```

```python
top_35_coords_final = pd.DataFrame([top_35_coords_lat, top_35_coords_lon], index=['lat', 'lon']).T
top_35_coords_final.head()
```

```python
top_35_coords_final.to_csv('assets/top_35_coords_final.csv')
```
